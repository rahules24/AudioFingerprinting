```python
import pandas as pd

songfile_name = "17. Climb" 
songfile_path = "logs/" + songfile_name  + "_hashes" +".txt"  

# Read the CSV file into a DataFrame
song_df = pd.read_csv(songfile_path)

# Convert DataFrame columns to lists
song_hashlist = song_df['hash'].tolist()
song_anchortimelist = song_df['anchor time'].tolist()
```

```python
import pandas as pd

samplefile_name = "17_Climb_1" 
samplefile_path = "logs/" + samplefile_name  + "_sample_hashes" +".txt"   

# Read the CSV file into a DataFrame
sample_df = pd.read_csv(samplefile_path)

# Convert DataFrame columns to lists
sample_hashlist = sample_df['hash'].tolist()
sample_anchortimelist = sample_df['anchor time'].tolist()
```

```python
import numpy as np
# Find indices of equal values
indices = []

# Creating a hash table (dictionary)
hash_table = {}

for i, value1 in enumerate(sample_hashlist):
    for j, value2 in enumerate(song_hashlist):
        if value1 == value2:
            key_offset = np.around(song_anchortimelist[j] - sample_anchortimelist[i], 2)

            if key_offset not in hash_table:
                hash_table[key_offset] = 1
            else:
                hash_table[key_offset] = hash_table[key_offset] + 1

            indices.append((i, j))

# Print the hash table
print(hash_table)

# Print the indices
print(indices)
```

```python
import matplotlib.pyplot as plt

# Separate x and y coordinates into separate lists
x_indices = [point[1] for point in indices]
y_indices = [point[0] for point in indices]
indexedx_list = [song_anchortimelist[i] for i in x_indices]
print(indexedx_list)

indexedy_list = [sample_anchortimelist[j] for j in y_indices]
print(indexedy_list)
# Create the scatter plot
plt.scatter(indexedx_list, indexedy_list)

# Set labels and title
plt.xlabel('time in full song')
plt.ylabel('time in sample')
plt.title('Scatter Plot')

# Increase the range of the plot
plt.xlim(0, 208)  # Set the x-axis limits from 0 to 208
plt.ylim(0, 12)  # Set the y-axis limits from 0 to 12
plt.subplots_adjust(left=0, right=5, top=5, bottom=0)

# Set equal aspect ratio
plt.gca().set_aspect('equal')

# Increase figure size
plt.figure(figsize=(20, 20))  # Set the figure size to 8x6 inches

# Display the plot
plt.show()
```

```python
keys = hash_table.keys()
values = hash_table.values()

plt.bar(keys, values)
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Hash Table Histogram')
plt.show()
```
